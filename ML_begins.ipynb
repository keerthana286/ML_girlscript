# **Data creation**
eqn `y=10x`
target->

``` 
x=[0,1,2,3,4,5,...]
y=[0,10,20,30,40,50,...]
```

````python
x=[i for i in range(21)]
print(x)
````

````python
y=[i for i in range(0,201,10)]
print(y)
````

# **ML approch**


````python
xtrain=x[:-5]
ytrain=y[:-5]

xtest=x[-5:]
ytest=y[-5:]
print(xtrain,"\n",ytrain,"\n",xtest,"\n",ytest)
````

````python
import tensorflow as tf
from tensorflow import keras
````

````python
model=tf.keras.Sequential([
                           tf.keras.layers.Dense(units=1,input_shape=[1])
                           ])
model.compile(optimizer='adam',loss='mae')
````

````python
model.fit(x=xtrain,y=ytrain,validation_data=(xtest,ytest),epochs=10000)
````
